```python
import numpy as np
import math
import itertools
```

ЗАВДАННЯ №1

```python
A = np.array([[1, 1, 1], [0.05, 0.07, 0], [0.05, 0, 0.06]])
x = np.array([50000, 2250, 1400])
result = np.linalg.solve(A, x)
print(f'На рахунку 1: {result[0]} у.о., На рахунку 2: {result[1]} у.о., На рахунку 3: {result[2]} у.о.')
```

ЗАВДАННЯ №5

```python
def vector(point_list):
    x_list = []
    for i in point_list:
        x_list.append(i[1])
    return np.array(x_list)


def matrix_from_points(point_list):
    A = []
    for p in point_list:
        a = []
        for i in range(len(point_list)):
            a.append(p[0]**i)
        a.reverse()
        A.append(a)
    return A
    
    
def fun(point_list):
    x = vector(point_list)
    print('масив х:')
    print(x)
    
    A = matrix_from_points(point_list)
    print('матриця А:')
    for a in A:
        print(a)
    
    result = np.linalg.solve(A, x)
    print('Коефіцієнти с:')
    print(result)
    
    
    
# Підставте свої точки в список кортежів А:   
    
A = [(1,12), (3,54), (-1,2)]

fun(A)
```
